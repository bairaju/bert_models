```python
import pandas as pd
from sklearn.model_selection import train_test_split
#read source data from csv file
df_train = pd.read_excel('E:/chartbot-bert/bert/dataset/filetosend1.xlsx', header = None)
df_test = pd.read_excel('E:/chartbot-bert/bert/dataset/test.xlsx',header= None)
df_test.columns = ['Class_label', 'Class_content']
df_train.columns = ['Class_label', 'Class_content']
#create a new dataframe for train, dev data
df_bert = pd.DataFrame({
    'label': df_train['Class_label'],
    'text': df_train['Class_content']})
#split into test, dev
df_bert_train, df_bert_dev = train_test_split(df_bert, test_size=0.01)
#create new dataframe for test data
df_bert_test = pd.DataFrame({
    'text': df_test['Class_content']})
#output tsv file, no header for train and dev
```

```python
df_bert_train.to_csv('E:/chartbot-bert/bert/dataset/train.tsv', sep='\t', index=False, header=False)
df_bert_dev.to_csv('E:/chartbot-bert/bert/dataset/dev.tsv', sep='\t', index=False, header=False)
df_bert_test.to_csv('E:/chartbot-bert/bert/dataset/test.tsv', sep='\t', index=False, header=True)
```
